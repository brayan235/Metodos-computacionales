```python
#Para convertir de C K
conv=273.15
metros=1e-2

lim=int(len(t0)/3.)

xs=[]
ys=[0.0*metros]
T0=[0.0+conv]
T1=[0.0+conv]
T2=[0.0+conv]
T3=[0.0+conv]
T4=[0.0+conv]

#seleccionando las x
for i in xrange(0,len(t0)):
    if i%3==0:
        xs.append(t0[i]*metros)
        
#seleccionando las y
for i in xrange(1,lim):
    ys.append(t0[3*i+1]*metros)
        
#seleccionando las temperaturas
for i in xrange(1,lim):
    T0.append(t0[3*i+2]+conv)
    T1.append(t1[3*i+2]+conv) 
    T2.append(t2[3*i+2]+conv) 
    T3.append(t3[3*i+2]+conv) 
    T4.append(t4[3*i+2]+conv)    
```

```python
#derivacion
```

```python
fac=0.01
k=34.0 #capacidad calorifica del material

dtdx0=[]
dtdy0=[]
dtdx1=[]
dtdy1=[]
dtdx2=[]
dtdy2=[]
dtdx3=[]
dtdy3=[]
dtdx4=[]
dtdy4=[]

#Calculando las derivadas a partir de: derivada = deltaT/delta(x o y) 
for i in xrange(0,len(xs)-1):
    #para t0
    dx0 = (T0[i+1]-T0[i]) / (xs[i+1]-xs[i]+fac)
    dy0 = (T0[i+1]-T0[i]) / (ys[i+1]-ys[i]+fac)
    dtdx0.append(dx0)
    dtdy0.append(dy0)
    
    #para t1
    dx1 = (T1[i+1]-T1[i]) / (xs[i+1]-xs[i]+fac)
    dy1 = (T1[i+1]-T1[i]) / (ys[i+1]-ys[i]+fac)
    dtdx1.append(dx1)
    dtdy1.append(dy1)
    
    #para t2
    dx2 = (T2[i+1]-T2[i]) / (xs[i+1]-xs[i]+fac)
    dy2 = (T2[i+1]-T2[i]) / (ys[i+1]-ys[i]+fac)
    dtdx2.append(dx2)
    dtdy2.append(dy2)
    
    #para t3
    dx3 = (T3[i+1]-T3[i]) / (xs[i+1]-xs[i]+fac)
    dy3 = (T3[i+1]-T3[i]) / (ys[i+1]-ys[i]+fac)
    dtdx3.append(dx3)
    dtdy3.append(dy3)
    
    #para t4
    dx4 = (T4[i+1]-T4[i]) / (xs[i+1]-xs[i]+fac)
    dy4 = (T4[i+1]-T4[i]) / (ys[i+1]-ys[i]+fac)
    dtdx4.append(dx4)
    dtdy4.append(dy4)

#Q:
q0=[]
q1=[]
q2=[]
q3=[]
q4=[]

#calculando el calor
for i in xrange(0,len(dtdx1)):
    q0.append( k*(dtdx0[i]**2+dtdy0[i]**2)**0.5 )
    q1.append( k*(dtdx1[i]**2+dtdy1[i]**2)**0.5 )
    q2.append( k*(dtdx2[i]**2+dtdy2[i]**2)**0.5 )
    q3.append( k*(dtdx3[i]**2+dtdy3[i]**2)**0.5 )
    q4.append( k*(dtdx4[i]**2+dtdy4[i]**2)**0.5 )
    
#print len(q0)
```
